```python
# Only selection datatype
df = df[df.datatype == 'Locomotion']
df.head()
```

```python
df = df.sort_values(by=['sn','pn','location','aname'])
df.head()
```

```python
df['time'] = df['stdate'] + " " + df['sttime']
df['time'].head()
```

```python
# mapping = dict(enumerate([2,5,6,8,12,16,26,32]))
# df['D'] = df['A'].map(mapping)
#loc_to_specie = { 'E': 'Erpobdella', 'G': 'Gammarus', 'R': 'Radix' }
mapping = lambda a : { 'E': 'Erpobdella', 'G': 'Gammarus', 'R': 'Radix' }[a[0]]
#df['specie'] = loc_to_specie[df.location.astype(str).str[0]].astype(str) #loc_to_specie[df['location'][0]]
df['specie'] = df['location'].map(mapping)
df['utime'] = pd.to_datetime(df['time']).astype('int64')//1e9
df['replica'] = df['animal']
df['timebin'] = df['end'] - df['start']
df['condition'] = u"%s_%s_%s" % (df['etude'],df['dose'],df['lot'])
df.head(17)
```
